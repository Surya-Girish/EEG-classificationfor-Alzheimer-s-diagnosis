```python
import numpy,pywt
from pyentrp import entropy
import math
from numpy import genfromtxt

def ar(a,b,c,d,e,f):
    a=numpy.array(a)
    b=numpy.array(b)
    c=numpy.array(c)
    d=numpy.array(d)
    e=numpy.array(e)
    f=numpy.array(f)
    return(a,b,c,d,e,f)

alpha=2
def renyi(X):
    data_set = list(set(X))
    freq_list = []
    for entry in data_set:
        counter = 0.
        for i in X:
            if i == entry:
                counter += 1
        freq_list.append(float(counter)/len(X))
    summation=0
    for freq in freq_list:
        summation+=math.pow(freq,alpha)
    Renyi_En=(1/float(1-alpha))*(math.log(summation,2))
    return(Renyi_En)

def energy(X):
    summ=0
    for i in X:
            summ+=(i**2)
    return(summ)

def Entropy(Y):
    n=[];d1=[];d2=[];d3=[];d4=[];a4=[];
    for i in range (Y.shape[0]):
        a=list(Y[i])
        coeff=pywt.wavedec(a,'db10',level=4)
        coeff=numpy.array(coeff)
        d1.append(coeff[0])
        d2.append(coeff[1])
        d3.append(coeff[2])
        d4.append(coeff[3])
        a4.append(coeff[4])
        n.append(Y[i])
    d1=numpy.array(d1)
    d2=numpy.array(d2)
    d3=numpy.array(d3)
    d4=numpy.array(d4)
    a4=numpy.array(a4)
    n=numpy.array(n)
    
        
    sample=[];shanon=[];permutation=[];reny=[];energ=[]
    for i in range(0,19):
        X=Y[i]
        std_X = numpy.std(X)
        ee=entropy.sample_entropy(X,2,0.2*std_X)
        sample.append(ee[0])
        shanon.append(entropy.shannon_entropy(X))
        energ.append(energy(X))
        permutation.append(entropy.permutation_entropy(X,3,1))
        reny.append(renyi(X))
    
    sample1=[];shanon1=[];permutation1=[];renyi1=[];energy1=[]
    for i in range(0,19):
        X=d1[i]
        std_X = numpy.std(X)
        ee=entropy.sample_entropy(X,2,0.2*std_X)
        sample1.append(ee[0])
        shanon1.append(entropy.shannon_entropy(X))
        energy1.append(energy(X))
        permutation1.append(entropy.permutation_entropy(X,3,1))
        renyi1.append(renyi(X))
    
    sample2=[];shanon2=[];permutation2=[];renyi2=[];energy2=[]
    for i in range(0,19):
        X=d2[i]
        std_X = numpy.std(X)
        ee=entropy.sample_entropy(X,2,0.2*std_X)
        sample2.append(ee[0])
        shanon2.append(entropy.shannon_entropy(X))
        energy2.append(energy(X))
        permutation2.append(entropy.permutation_entropy(X,3,1))
        renyi2.append(renyi(X))

    sample3=[];shanon3=[];permutation3=[];renyi3=[];energy3=[]
    for i in range(0,19):
        X=d3[i]
        std_X = numpy.std(X)
        ee=entropy.sample_entropy(X,2,0.2*std_X)
        sample3.append(ee[0])
        shanon3.append(entropy.shannon_entropy(X))
        energy3.append(energy(X))
        permutation3.append(entropy.permutation_entropy(X,3,1))
        renyi3.append(renyi(X))

    sample4=[];shanon4=[];permutation4=[];renyi4=[];energy4=[]
    for i in range(0,19):
        X=d4[i]
        std_X = numpy.std(X)
        ee=entropy.sample_entropy(X,2,0.2*std_X)
        sample4.append(ee[0])
        shanon4.append(entropy.shannon_entropy(X))
        energy4.append(energy(X))
        permutation4.append(entropy.permutation_entropy(X,3,1))
        renyi4.append(renyi(X))
    
    sample5=[];shanon5=[];permutation5=[];renyi5=[];energy5=[]
    for i in range(0,19):
        X=a4[i]
        std_X = numpy.std(X)
        ee=entropy.sample_entropy(X,2,0.2*std_X)
        sample5.append(ee[0])
        shanon5.append(entropy.shannon_entropy(X))
        energy5.append(energy(X))
        permutation5.append(entropy.permutation_entropy(X,3,1))
        renyi5.append(renyi(X))
    D1=[sample1,shanon1,permutation1,renyi1,energy1]
    D2=[sample2,shanon2,permutation2,renyi2,energy2]
    D3=[sample3,shanon3,permutation3,renyi3,energy3]
    D4=[sample4,shanon4,permutation4,renyi4,energy4]
    A4=[sample5,shanon5,permutation5,renyi5,energy5]
    N=[sample,shanon,permutation,reny,energ]
    return(N,D1,D2,D3,D4,A4)
X1= genfromtxt('eeg20o1.csv',delimiter=',',skip_header=0)
X2= genfromtxt('eeg21o1.csv', delimiter=',', skip_header=0)
X3= genfromtxt('eeg22o1.csv', delimiter=',', skip_header=0)
X4= genfromtxt('eeg23o1.csv', delimiter=',', skip_header=0)
X5= genfromtxt('eeg24o1.csv', delimiter=',', skip_header=0)
X6= genfromtxt('eeg25o1.csv', delimiter=',', skip_header=0)
X7= genfromtxt('eeg27o1.csv', delimiter=',', skip_header=0)
X8= genfromtxt('eeg28o1.csv', delimiter=',', skip_header=0)
X9= genfromtxt('eeg29o1.csv', delimiter=',', skip_header=0)
X10= genfromtxt('eeg32o1.csv', delimiter=',', skip_header=0)
X11= genfromtxt('eeg33o1.csv', delimiter=',', skip_header=0)
X12= genfromtxt('eeg35o1.csv', delimiter=',', skip_header=0)


N1,D11,D12,D13,D14,A14=Entropy(X1.T)
N2,D21,D22,D23,D24,A24=Entropy(X2.T)
N3,D31,D32,D33,D34,A34=Entropy(X3.T)
N4,D41,D42,D43,D44,A44=Entropy(X4.T)
N5,D51,D52,D53,D54,A54=Entropy(X5.T)
N6,D61,D62,D63,D64,A64=Entropy(X6.T)
N7,D71,D72,D73,D74,A74=Entropy(X7.T)
N8,D81,D82,D83,D84,A84=Entropy(X8.T)
N9,D91,D92,D93,D94,A94=Entropy(X9.T)
N10,D101,D102,D103,D104,A104=Entropy(X10.T)
N11,D111,D112,D113,D114,A114=Entropy(X11.T)
N12,D121,D122,D123,D124,A124=Entropy(X12.T)

N1,D11,D12,D13,D14,A14=ar(N1,D11,D12,D13,D14,A14)
N2,D21,D22,D23,D24,A24=ar(N2,D21,D22,D23,D24,A24)
N3,D31,D32,D33,D34,A34=ar(N3,D31,D32,D33,D34,A34)
N4,D41,D42,D43,D44,A44=ar(N4,D41,D42,D43,D44,A44)
N5,D51,D52,D53,D54,A54=ar(N5,D51,D52,D53,D54,A54)
N6,D61,D62,D63,D64,A64=ar(N6,D61,D62,D63,D64,A64)
N7,D71,D72,D73,D74,A74=ar(N7,D71,D72,D73,D74,A74)
N8,D81,D82,D83,D84,A84=ar(N8,D81,D82,D83,D84,A84)
N9,D91,D92,D93,D94,A94=ar(N9,D91,D92,D93,D94,A94)
N10,D101,D102,D103,D104,A104=ar(N10,D101,D102,D103,D104,A104)
N11,D111,D112,D113,D114,A114=ar(N11,D111,D112,D113,D114,A114)
N12,D121,D122,D123,D124,A124=ar(N12,D121,D122,D123,D124,A124)

y1=numpy.concatenate((N1,N2,N3,N4,N5,N6,N7,N8,N9,N10,N11,N12),axis=1)
y1=numpy.array(y1.T)

x1=numpy.concatenate((D11,D21,D31,D41,D51,D61,D71,D81,D91,D101,D111,D121),axis=1)
x1=numpy.array(x1.T)

x2=numpy.concatenate((D12,D22,D32,D42,D52,D62,D72,D82,D92,D102,D112,D122),axis=1)
x2=numpy.array(x2.T)

x3=numpy.concatenate((D13,D23,D33,D43,D53,D63,D73,D83,D93,D103,D113,D123),axis=1)
x3=numpy.array(x3.T)

x4=numpy.concatenate((D14,D24,D34,D44,D54,D64,D74,D84,D94,D104,D114,D124),axis=1)
x4=numpy.array(x4.T)

x5=numpy.concatenate((A14,A24,A34,A44,A54,A64,A74,A84,A94,A104,A114,A124),axis=1)
x5=numpy.array(x5.T)

X=numpy.concatenate((y1,x1,x2,x3,x4,x5),axis=1)
X=numpy.array(X)

numpy.savetxt("ADecomposedWaveFeatures.csv",X,delimiter=',')

```

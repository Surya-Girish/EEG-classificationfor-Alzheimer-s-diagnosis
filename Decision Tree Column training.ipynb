```python
def round_dict(d,k):
    res = dict()
    for key in d:
        res[key]=round(d[key],k)
    return res
        


def parse(d):
    fin = dict()
    i=-1
    k=3
    for key in d:
        i=i+1
        fin[i]=round_dict(key,k)
    fin = pd.DataFrame(fin,index= None)
    return(fin)
 

    
print(x1['0'])
print((parse(x1['0'])))
```

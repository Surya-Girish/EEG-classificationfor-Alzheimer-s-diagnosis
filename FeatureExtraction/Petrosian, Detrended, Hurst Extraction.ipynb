```python

import numpy
from numpy import genfromtxt
from antropy import petrosian_fd as pfd 
from antropy import detrended_fluctuation as dfa
from hurst import compute_Hc as he

d1= genfromtxt('eeg20o1.csv', delimiter=',', skip_header=0)
d2= genfromtxt('eeg21o1.csv', delimiter=',', skip_header=0)
d3= genfromtxt('eeg22o1.csv', delimiter=',', skip_header=0)
d4= genfromtxt('eeg23o1.csv', delimiter=',', skip_header=0)
d5= genfromtxt('eeg24o1.csv', delimiter=',', skip_header=0)
d6= genfromtxt('eeg25o1.csv', delimiter=',', skip_header=0)
d7= genfromtxt('eeg27o1.csv', delimiter=',', skip_header=0)
d8= genfromtxt('eeg28o1.csv', delimiter=',', skip_header=0)
d9= genfromtxt('eeg29o1.csv', delimiter=',', skip_header=0)
d10= genfromtxt('eeg32o1.csv', delimiter=',', skip_header=0)
d11= genfromtxt('eeg33o1.csv', delimiter=',', skip_header=0)
d12= genfromtxt('eeg35o1.csv', delimiter=',', skip_header=0)

d1=d1.T
d2=d2.T
d3=d3.T
d4=d4.T
d5=d5.T
d6=d6.T
d7=d7.T
d8=d8.T
d9=d9.T
d10=d10.T
d11=d11.T
d12=d12.T

def petf(d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12):
    
    p1=[];p2=[];p3=[];p4=[];p5=[];p6=[];p7=[];p8=[];p9=[];p10=[];p11=[];p12=[];
    for i in range(0,19):
        p1.append(pfd(d1[i]))
    for i in range(0,19):
        p2.append(pfd(d2[i]))
    for i in range(0,19):
        p3.append(pfd(d3[i]))
    for i in range(0,19):
        p4.append(pfd(d4[i]))
    for i in range(0,19):
        p5.append(pfd(d5[i]))
    for i in range(0,19):
        p6.append(pfd(d6[i]))
    for i in range(0,19):
        p7.append(pfd(d7[i]))
    for i in range(0,19):
        p8.append(pfd(d8[i]))
    for i in range(0,19):
        p9.append(pfd(d9[i]))
    for i in range(0,19):
        p10.append(pfd(d10[i]))
    for i in range(0,19):
        p11.append(pfd(d11[i]))
    for i in range(0,19):
        p12.append(pfd(d12[i]))   
    return(p1,p2,p3,p4,p5,p6,p7,p8,p9,p10,p11,p12)

def defa(d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12):
    
    p1=[];p2=[];p3=[];p4=[];p5=[];p6=[];p7=[];p8=[];p9=[];p10=[];p11=[];p12=[];
    for i in range(0,19):
        p1.append(dfa(d1[i]))
    for i in range(0,19):
        p2.append(dfa(d2[i]))
    for i in range(0,19):
        p3.append(dfa(d3[i]))
    for i in range(0,19):
        p4.append(dfa(d4[i]))
    for i in range(0,19):
        p5.append(dfa(d5[i]))
    for i in range(0,19):
        p6.append(dfa(d6[i]))
    for i in range(0,19):
        p7.append(dfa(d7[i]))
    for i in range(0,19):
        p8.append(dfa(d8[i]))
    for i in range(0,19):
        p9.append(dfa(d9[i]))
    for i in range(0,19):
        p10.append(dfa(d10[i]))
    for i in range(0,19):
        p11.append(dfa(d11[i]))
    for i in range(0,19):
        p12.append(dfa(d12[i]))   
    return(p1,p2,p3,p4,p5,p6,p7,p8,p9,p10,p11,p12)

def huex(d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12):
    
    h1=[];h2=[];h3=[];h4=[];h5=[];h6=[];h7=[];h8=[];h9=[];h10=[];h11=[];h12=[];
    for i in range(0,19):
        h,z,b= he(d1[i])
        h1.append(h)
    for i in range(0,19):
        h,z,b= he(d2[i])
        h2.append(h)
    for i in range(0,19):
        h,z,b= he(d3[i])
        h3.append(h)
    for i in range(0,19):
        h,z,b= he(d4[i])
        h4.append(h)
    for i in range(0,19):
        h,z,b= he(d5[i])
        h5.append(h)
    for i in range(0,19):
        h,z,b= he(d6[i])
        h6.append(h)
    for i in range(0,19):
        h,z,b= he(d7[i])
        h7.append(h)
    for i in range(0,19):
        h,z,b= he(d8[i])
        h8.append(h)
    for i in range(0,19):
        h,z,b= he(d9[i])
        h9.append(h)
    for i in range(0,19):
        h,z,b= he(d10[i])
        h10.append(h)
    for i in range(0,19):
        h,z,b= he(d11[i])
        h11.append(h)
    for i in range(d12.shape[0]):
        h,z,b= he(d12[i])
        h12.append(h)
    
    return(h1,h2,h3,h4,h5,h6,h7,h8,h9,h10,h11,h12)

print("Extracting Petrosian Values")
p1,p2,p3,p4,p5,p6,p7,p8,p9,p10,p11,p12 = petf(d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12)
print("Extracting Detrended Values")
a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12 = defa(d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12)
print("Hurst Exponent")
h1,h2,h3,h4,h5,h6,h7,h8,h9,h10,h11,h12 = huex(d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12)

x1=[a1,p1,h1]
x2=[a2,p2,h2]
x3=[a3,p3,h3]
x4=[a4,p4,h4]
x5=[a5,p5,h5]
x6=[a6,p6,h6]
x7=[a7,p7,h7]
x8=[a8,p8,h8]
x9=[a9,p9,h9]
x10=[a10,p10,h10]
x11=[a11,p11,h11]
x12=[a12,p12,h12]
X=numpy.concatenate((x1,x2,x3,x4,x5,x6,x7,x8,x9,x10,x11,x12),axis=1)
X=numpy.array(X)
X=X.T
a = numpy.asarray(X)
print(a.shape)
display(a)
numpy.savetxt("ACAntropy Features.csv", a, delimiter=",")
```

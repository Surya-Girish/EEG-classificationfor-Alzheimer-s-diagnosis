```python
import numpy as np
from numpy import genfromtxt
def hjorth(a):
    first_deriv = np.diff(a)
    second_deriv = np.diff(a,2)

    var_zero = np.mean(a ** 2)
    var_d1 = np.mean(first_deriv ** 2)
    var_d2 = np.mean(second_deriv ** 2)

    activity = var_zero
    morbidity = np.sqrt(var_d1 / var_zero)
    complexity = np.sqrt(var_d2 / var_d1) / morbidity

    return activity, morbidity, complexity

d1= genfromtxt('eeg20o1.csv', delimiter=',', skip_header=0)
d2= genfromtxt('eeg21o1.csv', delimiter=',', skip_header=0)
d3= genfromtxt('eeg22o1.csv', delimiter=',', skip_header=0)
d4= genfromtxt('eeg23o1.csv', delimiter=',', skip_header=0)
d5= genfromtxt('eeg24o1.csv', delimiter=',', skip_header=0)
d6= genfromtxt('eeg25o1.csv', delimiter=',', skip_header=0)
d7= genfromtxt('eeg27o1.csv', delimiter=',', skip_header=0)
d8= genfromtxt('eeg28o1.csv', delimiter=',', skip_header=0)
d9= genfromtxt('eeg29o1.csv', delimiter=',', skip_header=0)
d10= genfromtxt('eeg32o1.csv', delimiter=',', skip_header=0)
d11= genfromtxt('eeg33o1.csv', delimiter=',', skip_header=0)
d12= genfromtxt('eeg35o1.csv', delimiter=',', skip_header=0)

def hjpa(d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12):
    
    a1=[];a2=[];a3=[];a4=[];a5=[];a6=[];a7=[];a8=[];a9=[];a10=[];a11=[];a12=[];
    m1=[];m2=[];m3=[];m4=[];m5=[];m6=[];m7=[];m8=[];m9=[];m10=[];m11=[];m12=[];
    c1=[];c2=[];c3=[];c4=[];c5=[];c6=[];c7=[];c8=[];c9=[];c10=[];c11=[];c12=[];
    
    for i in range(d1.shape[0]):
        a, m, c = hjorth(d1[i])
        a1.append(a)
        m1.append(m)
        c1.append(c)
    for i in range(d2.shape[0]):
        a, m, c = hjorth(d2[i])
        a2.append(a)
        m2.append(m)
        c2.append(c)
    for i in range(d3.shape[0]):
        a, m, c = hjorth(d3[i])
        a3.append(a)
        m3.append(m)
        c3.append(c)
    for i in range(d4.shape[0]):
        a, m, c = hjorth(d4[i])
        a4.append(a)
        m4.append(m)
        c4.append(c)
    for i in range(d5.shape[0]):
        a, m, c = hjorth(d5[i])
        a5.append(a)
        m5.append(m)
        c5.append(c)
    for i in range(d6.shape[0]):
        a, m, c = hjorth(d6[i])
        a6.append(a)
        m6.append(m)
        c6.append(c)
    for i in range(d7.shape[0]):
        a, m, c = hjorth(d7[i])
        a7.append(a)
        m7.append(m)
        c7.append(c)
    for i in range(d8.shape[0]):
        a, m, c = hjorth(d8[i])
        a8.append(a)
        m8.append(m)
        c8.append(c)
    for i in range(d9.shape[0]):
        a, m, c = hjorth(d9[i])
        a9.append(a)
        m9.append(m)
        c9.append(c)
    for i in range(d10.shape[0]):
        a, m, c = hjorth(d10[i])
        a10.append(a)
        m10.append(m)
        c10.append(c)
    for i in range(d11.shape[0]):
        a, m, c = hjorth(d11[i])
        a11.append(a)
        m11.append(m)
        c11.append(c)
    for i in range(d12.shape[0]):
        a, m, c = hjorth(d12[i])
        a12.append(a)
        m12.append(m)
        c12.append(c)
    return(a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,m1,m2,m3,m4,m5,m6,m7,m8,m9,m10,m11,m12,c1,c2,c3,c4,c5,c6,
           c7,c8,c9,c10,c11,c12)

d1=d1.T
d2=d2.T
d3=d3.T
d4=d4.T
d5=d5.T
d6=d6.T
d7=d7.T
d8=d8.T
d9=d9.T
d10=d10.T
d11=d11.T
d12=d12.T


print("Hjorth parameters extracting")
a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12,m1,m2,m3,m4,m5,m6,m7,m8,m9,m10,m11,m12,c1,c2,c3,c4,c5,c6,c7,c8,c9,c10,c11,c12= hjpa(d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12)

x1=[a1,m1,c1]
x2=[a2,m2,c2]
x3=[a3,m3,c3]
x4=[a4,m4,c4]
x5=[a5,m5,c5]
x6=[a6,m6,c6]
x7=[a7,m7,c7]
x8=[a8,m8,c8]
x9=[a9,m9,c9]
x10=[a10,m10,c10]
x11=[a11,m11,c11]
x12=[a12,m12,c12]
X=np.concatenate((x1,x2,x3,x4,x5,x6,x7,x8,x9,x10,x11,x12),axis=1)
X = np.array(X)

X=X.T
a =np.asarray(X)
print(a.shape)

np.savetxt("ACHjorthFeatureRows.csv", a, delimiter=",")
```

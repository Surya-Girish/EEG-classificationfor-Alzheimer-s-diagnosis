```python
import numpy as np
import math
from numpy import genfromtxt
import statistics as st
d1= genfromtxt('eeg20o1.csv', delimiter=',', skip_header=0)
d2= genfromtxt('eeg21o1.csv', delimiter=',', skip_header=0)
d3= genfromtxt('eeg22o1.csv', delimiter=',', skip_header=0)
d4= genfromtxt('eeg23o1.csv', delimiter=',', skip_header=0)
d5= genfromtxt('eeg24o1.csv', delimiter=',', skip_header=0)
d6= genfromtxt('eeg25o1.csv', delimiter=',', skip_header=0)
d7= genfromtxt('eeg27o1.csv', delimiter=',', skip_header=0)
d8= genfromtxt('eeg28o1.csv', delimiter=',', skip_header=0)
d9= genfromtxt('eeg29o1.csv', delimiter=',', skip_header=0)
d10= genfromtxt('eeg32o1.csv', delimiter=',', skip_header=0)
d11= genfromtxt('eeg33o1.csv', delimiter=',', skip_header=0)
d12= genfromtxt('eeg35o1.csv', delimiter=',', skip_header=0)

d1=d1.T
d2=d2.T
d3=d3.T
d4=d4.T
d5=d5.T
d6=d6.T
d7=d7.T
d8=d8.T
d9=d9.T
d10=d10.T
d11=d11.T
d12=d12.T

def medd(d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12):
    
    med1=[];med2=[];med3=[];med4=[];med5=[];med6=[];med7=[];med8=[];med9=[];med10=[];med11=[];med12=[];
    med1=(np.median(d1,axis=1))
    med2=(np.median(d2,axis=1))
    med3=(np.median(d3,axis=1))
    med4=(np.median(d4,axis=1))
    med5=(np.median(d5,axis=1))
    med6=(np.median(d6,axis=1))
    med7=(np.median(d7,axis=1))
    med8=(np.median(d8,axis=1))
    med9=(np.median(d9,axis=1))
    med10=(np.median(d10,axis=1))
    med11=(np.median(d11,axis=1))
    med12=(np.median(d12,axis=1))
    return(med1,med2,med3,med4,med5,med6,med7,med8,med9,med10,med11,med12)
          


def modd(d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12):
    m1=[];m2=[];m3=[];m4=[];m5=[];m6=[];m7=[];m8=[];m9=[];m10=[];m11=[];m12=[];
    for i in range(d1.shape[0]):
        m1.append(st.mode(d1[i]))
        m2.append(st.mode(d2[i]))
        m3.append(st.mode(d3[i]))
        m4.append(st.mode(d4[i]))
        m5.append(st.mode(d5[i]))
        m6.append(st.mode(d6[i]))
        m7.append(st.mode(d7[i]))
        m8.append(st.mode(d8[i]))
        m9.append(st.mode(d9[i]))
        m10.append(st.mode(d10[i]))
        m11.append(st.mode(d11[i]))
        m12.append(st.mode(d12[i]))
        
    return(m1,m2,m3,m4,m5,m6,m7,m8,m9,m10,m11,m12)

def qrt(d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12):
    qa1=[];qa2=[];qa3=[];qa4=[];qa5=[];qa6=[];qa7=[];qa8=[];qa9=[];qa10=[];qa11=[];qa12=[];
    qb1=[];qb2=[];qb3=[];qb4=[];qb5=[];qb6=[];qb7=[];qb8=[];qb9=[];qb10=[];qb11=[];qb12=[];
    qc1=[];qc2=[];qc3=[];qc4=[];qc5=[];qc6=[];qc7=[];qc8=[];qc9=[];qc10=[];qc11=[];qc12=[];
    
    q1=np.quantile(d1,.25,axis=1)
    q3=np.quantile(d1,.75,axis=1)
    iq=q3-q1
    qa1=q1
    qb1=q3
    qc1=iq
    q1=np.quantile(d2,.25,axis=1)
    q3=np.quantile(d2,.75,axis=1)
    iq=q3-q1
    qa2=q1
    qb2=q3
    qc2=iq
    q1=np.quantile(d3,.25,axis=1)
    q3=np.quantile(d3,.75,axis=1)
    iq=q3-q1
    qa3=q1
    qb3=q3
    qc3=iq
    q1=np.quantile(d4,.25,axis=1)
    q3=np.quantile(d4,.75,axis=1)
    iq=q3-q1
    qa4=q1
    qb4=q3
    qc4=iq
    q1=np.quantile(d5,.25,axis=1)
    q3=np.quantile(d5,.75,axis=1)
    iq=q3-q1
    qa5=q1
    qb5=q3
    qc5=iq
    q1=np.quantile(d6,.25,axis=1)
    q3=np.quantile(d6,.75,axis=1)
    iq=q3-q1
    qa6=q1
    qb6=q3
    qc6=iq
    q1=np.quantile(d7,.25,axis=1)
    q3=np.quantile(d7,.75,axis=1)
    iq=q3-q1
    qa7=q1
    qb7=q3
    qc7=iq
    q1=np.quantile(d8,.25,axis=1)
    q3=np.quantile(d8,.75,axis=1)
    iq=q3-q1
    qa8=q1
    qb8=q3
    qc8=iq
    q1=np.quantile(d9,.25,axis=1)
    q3=np.quantile(d9,.75,axis=1)
    iq=q3-q1
    qa9=q1
    qb9=q3
    qc9=iq
    q1=np.quantile(d10,.25,axis=1)
    q3=np.quantile(d10,.75,axis=1)
    iq=q3-q1
    qa10=q1
    qb10=q3
    qc10=iq
    q1=np.quantile(d11,.25,axis=1)
    q3=np.quantile(d11,.75,axis=1)
    iq=q3-q1
    qa11=q1
    qb11=q3
    qc11=iq
    q1=np.quantile(d12,.25,axis=1)
    q3=np.quantile(d12,.75,axis=1)
    iq=q3-q1
    qa12=q1
    qb12=q3
    qc12=iq
    return(qa1,qa2,qa3,qa4,qa5,qa6,qa7,qa8,qa9,qa10,qa11,qa12,qb1,qb2,qb3,qb4,qb5,qb6,qb7,qb8,qb9,qb10,qb11,
       qb12,qc1,qc2,qc3,qc4,qc5,qc6,qc7,qc8,qc9,qc10,qc11,qc12)

print("Quantile extraction")
qa1,qa2,qa3,qa4,qa5,qa6,qa7,qa8,qa9,qa10,qa11,qa12,qb1,qb2,qb3,qb4,qb5,qb6,qb7,qb8,qb9,qb10,qb11,qb12,qc1,qc2,qc3,qc4,qc5,qc6,qc7,qc8,qc9,qc10,qc11,qc12 = qrt(d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12)
print("Median extraction")
a1,a2,a3,a4,a5,a6,a7,a8,a9,a10,a11,a12= medd(d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12)
print("Mode extraction")
b1,b2,b3,b4,b5,b6,b7,b8,b9,b10,b11,b12 = modd(d1,d2,d3,d4,d5,d6,d7,d8,d9,d10,d11,d12)

x1=[a1,b1,qa1,qb1,qc1]
x2=[a2,b2,qa2,qb2,qc2]
x3=[a3,b3,qa3,qb3,qc3]
x4=[a4,b4,qa4,qb4,qc4]
x5=[a5,b5,qa5,qb5,qc5]
x6=[a6,b6,qa6,qb6,qc6]
x7=[a7,b7,qa7,qb7,qc7]
x8=[a8,b8,qa8,qb8,qc8]
x9=[a9,b9,qa9,qb9,qc9]
x10=[a10,b10,qa10,qb10,qc10]
x11=[a11,b11,qa11,qb11,qc11]
x12=[a12,b12,qa12,qb12,qc12]

X=np.concatenate((x1,x2,x3,x4,x5,x6,x7,x8,x9,x10,x11,x12),axis=1)

X=np.array(X)
X=X.T
a=np.asarray(X)
print(a.shape)
display(a)
np.savetxt("Mode+Quantiles.csv",a,delimiter=',')
```
